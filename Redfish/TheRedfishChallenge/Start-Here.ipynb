## Challenge description

The goal of this challenge is to retrieve the information provided in the `System Information --> Device Inventory` view of an iLO 5 with a minimum number of Redfish GET requests in order to automate as much processing as possible in  in the remotely managed Redfish engine. 

The following screenshot shows the `Device Inventory` list of a HPE DL360 Gen10 server and the properties you have to retrieve: `Location`, `Product Name`, `Product Version`, `Firmware Version`, `Status`.

![Device Inventory from iLO GUI](Pictures/DeviceInventoryFromiLOGUI.png)

When judging the winner of this challenge, int the case of a tie, session authenticated programs will receive a higher score. Programs implementing good Redfish practices (meaning crawling the Redfish tree to find the resources path instead of assuming a definite location) will also receive a higher score.

For example, a program discovering the list and the names of all the chassis in a server will get a higher score than a program assuming that the first chassis is called `1`, the second `2`...

The formating of the output is not an important criteria; a raw JSON output is fine.

## Clone the repo in your Jupyter environment 

Once you have a copy of the repo in your own GitHub account, **you have to clone** it into this Jupyter environment. Here are the basic steps required to achieve this clone:

1. Copy the Redfish GitHub repo location in **your clipboard**

![ClipboardGitLocation](Pictures/ClipboardOfGit.png)

2. From your Jupyter environment, **Select `Terminal`** at the bottom left of the main `Launcher` pane, . If you don't have any Launcher pane, click on the `+` sign in the upper left of the left pane.

![LauncherPane](Pictures/StartTerminal.png)

3. Perform the **clone**

Issue the following commands in your terminal, using Ctlr-V to paste the content of your clipboard: 

```Bash
cd Discover/TheRedfishChallenge
git clone https://github.com/<YourGitHubLocation>/RedfishChallenge.git
```
